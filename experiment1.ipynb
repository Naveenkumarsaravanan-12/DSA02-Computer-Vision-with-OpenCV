```python
import cv2
import pandas as pd
import numpy as np
import  matplotlib.pyplot as plt
```

```python
img=cv2.imread('inputa.jpg',0)
plt.imshow(img,cmap='gray')
plt.title("GrayScale Image")
plt.axis('off')
plt.show()
```

```python
img1=cv2.imread('inputa.jpg',1)
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
blur=cv2.GaussianBlur(img1,(15,15),0)
plt.imshow(blur)
plt.title("GAUSSIAN BLUR  Image")
plt.axis('off')
plt.show()
```

```python
img2=cv2.imread('input.jpg',1)
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
canny=cv2.Canny(img2,100,200)
plt.imshow(canny,cmap='gray')
plt.title("Canny  Edage Image")
plt.axis('off')
plt.show()
```

```python
img3=cv2.imread('input.jpg',0)
canny=cv2.Canny(img3,100,200)
kernal=np.ones((5,5),np.uint8)
dilate=cv2.dilate(canny,kernal,iterations=1)
plt.imshow(dilate,cmap='gray')
plt.title("Dilated Image")
plt.axis('off')
plt.show()

```

```python
import cv2
img4=cv2.imread('input.jpg',0)
canny=cv2.Canny(img4,100,200)
kernal=np.ones((5,5),np.uint8)
dilate=cv2.dilate(canny,kernal,iterations=1)
erode=cv2.erode(dilate,kernal,iterations=1)
plt.imshow(erode,cmap='gray')
plt.title("Eroded Image")
plt.axis('off')
plt.show()
```
